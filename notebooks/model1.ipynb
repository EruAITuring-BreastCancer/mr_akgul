```python
from google.colab import drive
import os

drive.mount('/content/drive')

folder_path = '/content/drive/MyDrive/Meme_Kanseri_Proje_Verileri'

if os.path.exists(folder_path):
    for root, dirs, files in os.walk(folder_path):
        level = root.replace(folder_path, '').count(os.sep)
        indent = ' ' * 4 * level
        print(f'{indent}{os.path.basename(root)}/')
        subindent = ' ' * 4 * (level + 1)

        for f in files[:5]:
            print(f'{subindent}{f}')

        if len(files) > 5:
            print(f'{subindent}... ve {len(files) - 5} dosya daha')
else:
    print("Belirtilen klasör yolu bulunamadı.")
```

```python
import pandas as pd
import glob

csv_files = glob.glob(folder_path + '/**/*.csv', recursive=True)

if not csv_files:
    print("Klasörde CSV dosyası bulunamadı.")
else:
    for file in csv_files:
        print(f"\nİncelenen Dosya: {file}")
        try:
            df = pd.read_csv(file)
            print(df.info())
            print("\nÖrnek Veri:")
            print(df.head())

            birads_cols = [col for col in df.columns if 'birads' in col.lower() or 'bi-rads' in col.lower()]

            if birads_cols:
                for col in birads_cols:
                    print(f"\n{col} Sınıf Dağılımı:")
                    print(df[col].value_counts(dropna=False))
            else:
                print("\nDosyada BI-RADS sütunu tespit edilemedi.")
        except Exception as e:
            print(f"Dosya okunurken hata oluştu: {e}")
```
