```python
import pandas as pd
import os

# Yollar
base_path = '/content/drive/MyDrive/Meme_Kanseri_Proje_Verileri/'
split_path = os.path.join(base_path, 'SPLIT_DATA/')
folder_path = os.path.join(base_path, 'RSNA_Cropped/')

# Klasördeki gerçek CSV dosyalarını listele
available_csvs = [f for f in os.listdir(split_path) if f.endswith('.csv')]
print(f"SPLIT_DATA içindeki dosyalar: {available_csvs}\n")

print(f"{'Dosya Adı':<25} | {'Kayıt':<8} | {'Eşleşen':<8} | {'Eksik':<6} | {'Sızıntı'}")
print("-" * 75)

all_patients = {}

for file_name in available_csvs:
    path = os.path.join(split_path, file_name)
    df = pd.read_csv(path)

    # Dosya adlarını oluştur
    df['filename'] = df['patient_id'].astype(str) + '_' + df['image_id'].astype(str) + '.png'

    # Fiziksel dosya kontrolü
    matches = df['filename'].apply(lambda x: os.path.exists(os.path.join(folder_path, x))).sum()
    missing = len(df) - matches

    # Sızıntı kontrolü
    current_patients = set(df['patient_id'].unique())
    leakage = 0
    for other_patients in all_patients.values():
        leakage += len(current_patients.intersection(other_patients))

    all_patients[file_name] = current_patients

    print(f"{file_name:<25} | {len(df):<8} | {matches:<8} | {missing:<6} | {leakage}")
```

```python
import os
import pandas as pd
import torch
from torch.utils.data import Dataset
from PIL import Image

# 1. Klasör Yolunu Tanımla
image_dir = '/content/drive/MyDrive/Meme_Kanseri_Proje_Verileri/VinDr_Mammo/'

# 2. Dosya İsimlerinden Veri Seti Oluştur
data_list = []
for file_name in os.listdir(image_dir):
    if file_name.endswith('.png'):
        # Dosya isminin başındaki sayıyı al (Örn: '9_Unnamed...' -> label: 9)
        try:
            label = int(file_name.split('_')[0])
            data_list.append({
                'file_path': os.path.join(image_dir, file_name),
                'label': label
            })
        except ValueError:
            continue

df = pd.DataFrame(data_list)

# Veri özeti
print(f"Toplam Görüntü Sayısı: {len(df)}")
print("Etiket Dağılımı:\n", df['label'].value_counts())

# 3. PyTorch Dataset Sınıfı
class MammoDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.df = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_path = self.df.iloc[idx]['file_path']
        image = Image.open(img_path).convert('RGB')
        label = self.df.iloc[idx]['label']

        # Etiketleri 0'dan başlayan indexlere çevirmek gerekebilir (örn: 1-14 arası ise)
        # label = label - 1

        if self.transform:
            image = self.transform(image)

        return image, label

print("\n--- Model eğitimi için Dataset hazır ---")
```
